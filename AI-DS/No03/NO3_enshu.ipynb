### ■問題3-2 不偏分散のシミュレーション

1)０と１をnumpy.random.randint(0,2) を使って10000個ランダムに発生させてください。これを母集団 pop とします。<br>
2)この母集団の分散を標本分散として計算してください。これを母分散といいます。<br>
3)母集団から100個の標本を np.random.choiceを使ってランダムに選択して①標本分散②不偏分散を計算してください<br>
4)標本のサンプリングを50回繰り返してデータを蓄積してください<br>
5)①標本分散②不偏分散のそれぞれの平均を計算してください<br>

【考察】<br>
プログラム実行を何回か繰り返してください<br>
(1)標本分散と不偏分散について母分散との比較をしてください。どちらが母分散に近いでしょうか？<br>
(2)母集団から小さい標本数のサンプリングをして母分散の推定をする場合、①標本分散②不偏分散のいずれが適当でしょうか？<br>


```python
#1) 母集団popを発生
kaisu = 10000
pop =[]
#ここにコードを2行追加
pop = np.random.randint(0,2, size=kaisu)

#2)母分散
print(f"母分散  = {np.var(pop, ddof = 0):.3f}")

#3)重複なしに100個の標本データをサンプリング ①標本分散 ②不偏分散を計算
#4)これを50回繰り返して①標本分散 ②不偏分散のデータを蓄積
sample_var  = []
unbias_var  = []
sampling_count = 50
sample_num = 100
for j in range(sampling_count):
    #ループ内にコードを3行追加
    sample = np.random.choice(pop, size=sample_num, replace=False)
    sample_var.append(np.var(sample, ddof = 0)) # 標本分散
    unbias_var.append(np.var(sample, ddof = 1)) # 不偏分散
    
    

#5)①標本分散②不偏分散のそれぞれの平均を計算
print(f"100サンプルの標本分散の平均 = { np.mean(sample_var):.3f}")
print(f"100サンプルの不偏分散の平均 = { np.mean(unbias_var):.3f}")    
```

#【考察】
#ここに考察の解答を書く

母分散が0.250であることを考えると、不偏分散の平均が0.250であるので不偏分散がより正確であると言える。
標本分散では、サンプリング結果をnで割るのでサンプリングを過小評価してしまうが、不偏分散ではサンプリング結果をn-1で割るためサンプリングを過小評価しないので不偏分散の方が母分散の推定としては適当だと考えられる。
